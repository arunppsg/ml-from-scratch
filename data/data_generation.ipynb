```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

### Generating data for classification

```python
def generate_classification_data(mu0, mu1, cov0, cov1):
    y = np.random.binomial(1, 0.5, size=1000)
    x = []
    for label in y:
        if label == 0:
            x.append(np.random.multivariate_normal(mu0, cov0, 1)[0])
        else:
            x.append(np.random.multivariate_normal(mu1, cov1, 1)[0])

    x = np.array(x)
    data = pd.DataFrame({'x1':x.T[0], 'x2':x.T[1], 'y':y})
    return data
```

```python
mu0 = [-2, -2]
mu1 = [2, 2]
cov0 = np.eye(2) * 0.5
cov1 = np.eye(2) * 0.5

data = generate_classification_data(mu0, mu1, cov0, cov1)
data.to_csv('classification_data.csv', index=False)


plt.figure(figsize=(6, 6))
plt.scatter(data['x1'], data['x2'], c=data['y'])
```

### Generating data for regression

```python
def generate_regression_data():
    coefs = np.array([2, -5, 8])
    x1 = np.random.uniform(-10, 10, 1000)
    x2 = np.random.uniform(-5, 5, 1000)
    x3 = np.random.uniform(-20, 20, 1000)
    noise = np.random.normal(20, 20, 1000)
    x = np.array([x1, x2, x3])
    y = np.sum(x.T * coefs, axis=1) + noise
    df = pd.DataFrame({'x1':x1, 'x2':x2, 'x3':x3, 'y':y})
    return df
```

```python
df = generate_regression_data()

df.to_csv('regression_data.csv', index=False)
```
